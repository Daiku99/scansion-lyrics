```python
import regex as re

def contains_japanese_or_korean(text):
    # Regular expression pattern for Japanese and Korean characters
    pattern = re.compile(r'[\p{Script=Hiragana}\p{Script=Katakana}\p{Script=Hangul}]+')
    
    # Search for the pattern in the text
    match = pattern.search(text)
    
    # If a match is found, return True; otherwise, return False
    return match is not None

# Example usage:
text1 = "こんにちは 안녕하세요"
text2 = "Hello, World!"
text3 = "今天"

print(contains_japanese_or_korean(text1))  # True
print(contains_japanese_or_korean(text3))  # False

def remove_punctuation(string):
    # initializing punctuations string
    punc = '''!()-[]{};:'"\,./?@#$%^&*_~。，！∶（）【】？“”；�'''
    # Removing punctuations in string
    # Using loop + punctuation string
    for ele in string:
        if ele in punc:
            string = string.replace(ele, "")
    return string

#training_data
data = list()

with open("./three_height_scansion_tones_poem_clean.csv","w") as fout:
    with open("./three_height_scansion_tones_poem.csv","r") as fin:
        lines = [line.strip() for line in fin.readlines()]
        for sent in lines[1:]:
            sent = sent.split("\t")
            keyword,wb, tones, lyric = sent[0],sent[1],sent[2],sent[3]
            tmp = str()
            if len(keyword) == 0:
                print("0 keywords")
            else:
                lyr = remove_punctuation(lyric)
                if 1 < len(lyr) < 20:
                    if contains_japanese_or_korean(lyr):
                        pass
                    else:
                        if lyr[0] in ["女","男","合"]:
                            if lyr[1] in ["人","孩","卺","儿","娲","生"]:
                                # data.append(lyr)
                                tmp = lyr
                            else:
                                data.append(lyr[1:])
                        elif lyr[1] in ["女","男","合"]:
                            if "男女老少"  in lyr:
                                # data.append(lyr)
                                tmp = lyr
                        else:
                            # data.append(lyr)
                            tmp = lyr
            if len(tmp)!=0 :
                d = keyword + "\t" + wb + "\t" + tones + "\t" + lyric +"\n" 
                fout.write(d)

                    
```

```python
import jieba
from pypinyin import lazy_pinyin, Style

sentence = "我喜欢使用jieba进行中文分词"
seg_list = jieba.cut(sentence, cut_all=False)
segmented_sentence = " ".join(seg_list)

print(segmented_sentence)


word = "天文台"
pinyin_list = lazy_pinyin(word, style=Style.TONE3, tone_sandhi=True)

pinyin_numeric_tones = " ".join(pinyin_list)

print(pinyin_numeric_tones)#wrong conversion
print([x[-1] for x in pinyin_numeric_tones.split()])


def pinyin_automatic(lst):
    res = list()
    for wd in lst:
        pinyin_list = lazy_pinyin(wd, style=Style.TONE3, tone_sandhi=True,neutral_tone_with_five=True)

        pinyin_numeric_tones = " ".join(pinyin_list)
        tmp = [x[-1] for x in pinyin_numeric_tones.split()]
        if len(tmp) == 3:
            tmp = conversion_rule(tmp)
        elif len(tmp) == 2:
            tmp = conversion_rule(tmp)
        res.append(tmp)

    return res



# automatic package conversion is not accurate
'''
Rule 1: Tone 3 + 3 -> 2 + 3
Rule 2: Tone 1 + 2 + 2 -> 1 + 1 + 2
Rule 3: Tone 2 + 2 + 4 -> 2 + 1 + 4

'''

def conversion_rule(segmented):
    if len(segmented) == 3:
        if segmented == ["1","2","2"]:
            return ["1","1","2"]
        elif segmented == ["2","2","4"]:
            return ["2","1","4"]
        else:
            return segmented
    elif len(segmented) == 2:
        if segmented[0] == "3":
            if segmented[-1] == "3":
                segmented[0] = "2"
                return ["2","3"]
        # elif segmented[0] == "1":
        #     if segmented[-1] == "1":
        #         segmented[0] == "-1"
        #         return ["-1","1"]
        return segmented

    else:
        return segmented

def segmentation(string):
    seg = jieba.cut(string, cut_all=False)
    res = " ".join(seg)
    return res.split()

def repeat_string(s, n):
    return s * n


def boundary_seq_ie(lst):
    res = str()
    for word in lst:
        if len(word) == 1:
            res += "<b0> "#word with only one Hanzi
        else:
            res = res + repeat_string("<b0> ",len(word)-1) + "<b1> "#多字词，除了最后一个字都是<b0>
    return res

def lst_2_tone_num(lst):
    res = list()
    for i in lst:
        for ii in i:
            try:
                res.append(int(ii))
            except ValueError:
                continue
        
    return res
```

```python
with open("three_height_scansion_tones_poem.csv","w") as fout:
    mapping = {1:"<h>",2:"<m>",3:"<l>",4:"<h>",5:"<l>"}
    for sent in keyword_sent:
        keyword,sent = sent[0],sent[-1]
        py = pinyin_automatic(segmentation(sent))
        try:
            for idx in range(len(py)):
                py[idx] = conversion_rule(py[idx])
                #                 tones = lst_2_tone_num(pinyin_automatic(segmentation(lyr)))
                            #             print(tones)
                tones = lst_2_tone_num(py)
                for i in range(len(tones)):
                    tones[i] = mapping[tones[i]]
                wb = boundary_seq_ie(segmentation(sent))
            if len(keyword) < len(sent):
                d = keyword + "\t" + wb + "\t" + " ".join(tones) + "\t" + sent +"\n" 
                fout.write(d)
        except KeyError:
            print(sent)
            continue
```
