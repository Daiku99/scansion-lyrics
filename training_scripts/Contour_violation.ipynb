```python
def open_gpt2(path):
    if "3" in path:
        with open(path,"r") as fin:
            return [line.strip().split("\t")[0].split() for line in fin.readlines()[1:]]
    else:
        five_height = {"<t1>": 4, "<t2>": 3, "<t3>": 1,"<t4>":5,"<t0>":0}
        output = list()
        with open(path,"r") as fin:
            for line in fin.readlines()[1:]:
                line = line.strip().split("\t")[0].split()
                res = list()
                for sym in line:
                    res.append(five_height[sym])
                output.append(res)   
        return output
```

```python
def open_3_height(path):
    with open(path,"r") as fin:
        return [line.strip().split() for line in fin.readlines()]
    
def open_neume_5(path):
    five_height = {"<t1>": 4, "<t2>": 3, "<t3>": 1,"<t4>":5,"<t0>":0}
    output = list()
    with open(path,"r") as fin:
        for line in fin.readlines():
            line = line.strip().split()
            res = list()
            for sym in line:
                res.append(five_height[sym])
            output.append(res)   
    return output

open_neume_5(root_path + outputs_scansion[-3])
```

```python
def is_first_element_number(lst):
    if len(lst) > 0:
        first_element = lst[0]
        return isinstance(first_element, (int, float))
    return False

# Test cases
input1 = [1, 2, 3, 4]
input2 = ["Hello", "World"]
input3 = [3.14, 2.718, 1.618]
input4 = []

print(is_first_element_number(input1))  # True
print(is_first_element_number(input2))  # False
print(is_first_element_number(input3))  # True
print(is_first_element_number(input4))  # False


def is_first_element_3_height(lst):
    if len(lst) > 0:
        first_element = lst[0]
#         print(first_element)
        try:
            if first_element[1] in ["h","m","l"]:
                return True
        except:
            return False
    return False

# Test cases
input1 = [1, 2, 3, 4]
input2 = ["<t>","<h>"]
input3 = [3.14, 2.718, 1.618]
input4 = ["<h>","<l>"]

print(is_first_element_3_height(input1))  # True
print(is_first_element_3_height(input2))  # False
print(is_first_element_3_height(input3))  # True
print(is_first_element_3_height(input4))  # False
```

```python
def generate_contour_symbols(lst):
    init_res = ["<flat>"]
    for i in range(1,len(lst)):
        if lst[i] - lst[i-1] > 0 :
            init_res.append("<rise>")
        elif lst[i] == lst[i-1]:
            init_res.append("<flat>")
        else:
            init_res.append("<fall>")
    return init_res

def convert_symbols(lst):# absolute pitch, 3-height scansion & 5-height scansion
    three_height_mapping = {"<h>":5,"<m>":3,"<l>":0}
    five_height_mapping = {"<t1>": 4, "<t2>": 3, "<t3>": 1,"<t4>":5,"<t0>":0}
    if is_first_element_number(lst):
        res = generate_contour_symbols(lst)
    elif is_first_element_3_height(lst):        
        numbers = convert_symbols_to_numbers(lst,three_height_mapping)
        res = generate_contour_symbols(numbers)
    else:
        numbers = convert_symbols_to_numbers(lst,five_height_mapping)
        res = generate_contour_symbols(numbers)        
            
    return res


def convert_symbols_to_numbers(symbol_list, symbol_to_number):
    # Initialize an empty list to store the converted numbers
    numbers = []
    
    # Iterate through the symbol list
    for symbol in symbol_list:
        # Use the dictionary to look up the corresponding number
        if symbol in symbol_to_number:
            number = symbol_to_number[symbol]
            numbers.append(number)
        else:
            # Handle the case where the symbol is not found in the dictionary
            # You can raise an error, set a default value, or handle it as needed
            raise ValueError(f"Symbol '{symbol}' not found in the dictionary.")
    
    return numbers
```

```python
def chech_violation(m_scansion,l_scansion):
    m_scansion, l_scansion = convert_symbols(m_scansion), convert_symbols(l_scansion)
    if len(m_scansion) == len(l_scansion):
        init_score = 0
        for i in range(len(m_scansion)):
            #  melody scansion, lyrics scansion's symbol -> <flat>, accepted
            if m_scansion[i] == "<flat>":
                pass
            elif l_scansion[i] == "<flat>":
                pass
            elif m_scansion[i] != l_scansion[i]:
                # only <rise> VS <fall>
                init_score -= 1
        return init_score
    else:
        return False
    
def check_correction(m_scansion,l_scansion):
    m_scansion, l_scansion = convert_symbols(m_scansion), convert_symbols(l_scansion)
    if len(m_scansion) == len(l_scansion):
        init_score = 0
        for i in range(len(m_scansion)):
            if m_scansion[i] == l_scansion[i]:
                init_score += 1
        return init_score
    else:
        return False
            
def joint_contour_score(m_scansion,l_scansion):
    i,j = check_correction(m_scansion,l_scansion), chech_violation(m_scansion,l_scansion)
    if isinstance(i, int) and isinstance(j, int) :
        print("Correction: ",i, "Violation: ", j)
        return i + j
    else:
        print("No score for this parallel scansion!")
        return False

        
```
